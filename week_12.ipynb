# RNN

```python
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense
import numpy as np

# Define vocabulary
vocab = ['d', 'e', 'p', '<stop>']
char2idx = {c: i for i, c in enumerate(vocab)}
idx2char = {i: c for c, i in char2idx.items()}

# Example input sequence: "dep"
sequence = ['d', 'e', 'p']
target = ['e', 'p', '<stop>']  # Next characters

# One-hot encode input and target
X = np.eye(len(vocab))[[char2idx[c] for c in sequence]].reshape(len(sequence), 1, len(vocab))
y = np.eye(len(vocab))[[char2idx[c] for c in target]]
# Build RNN model
model = Sequential([
    SimpleRNN(8, input_shape=(1, len(vocab))),
    Dense(len(vocab), activation='softmax')  # softmax output
])

model.compile(loss='categorical_crossentropy', optimizer='adam')  # cross-entropy loss

# Train the model
model.fit(X, y, epochs=1000, verbose=0)

# Predict next character after 'p'
test = np.eye(len(vocab))[char2idx['p']].reshape(1, 1, len(vocab))
pred = model.predict(test)
print("Next char prediction:", idx2char[np.argmax(pred)])
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, Flatten
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# 1. Prepare the data
reviews = ["good movie", "bad movie", "excellent film", "terrible acting"]
labels = np.array([1, 0, 1, 0]) # 1 for positive, 0 for negative

# Tokenize the text
tokenizer = Tokenizer(num_words=None) # Consider all words
tokenizer.fit_on_texts(reviews)
word_index = tokenizer.word_index
vocab_size = len(word_index) + 1 # +1 for the padding token

# Convert text to sequences of integers
sequences = tokenizer.texts_to_sequences(reviews)

# Pad sequences to have the same length
max_length = max(len(seq) for seq in sequences)
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post')

# 2. Define the Keras model
embedding_dim = 10
model = Sequential([
    Embedding(vocab_size, embedding_dim, input_length=max_length), # Embed word indices into vectors
    SimpleRNN(8),                                              # Simple RNN layer with 8 hidden units
    Dense(1, activation='sigmoid')                             # Output layer with sigmoid for binary classification
])

# 3. Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 4. Train the model (for a few epochs for demonstration)
model.fit(padded_sequences, labels, epochs=10, verbose=0)

# 5. Predict the sentiment of the first review
test_review = ["amazing movie"]
test_sequence = tokenizer.texts_to_sequences(test_review)
padded_test_sequence = pad_sequences(test_sequence, maxlen=max_length, padding='post')
prediction = model.predict(padded_test_sequence)
sentiment = "Positive" if prediction[0] > 0.5 else "Negative"

# 6. Print the result
print(f"Review: '{test_review[0]}', Predicted Sentiment: {sentiment} (Confidence: {prediction[0][0]:.2f})")

```

# LSTM

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense
import numpy as np

# Vocabulary and word-index mapping
vocab = {'the': 0, 'sun': 1, 'is': 2, 'shining': 3, 'bright': 4, 'today': 5}
idx2word = {i: w for w, i in vocab.items()}
vocab_size = len(vocab)

# Input sequence: "the sun is shining"
X = np.array([[0, 1, 2, 3]])  # indices of words
y = np.array([4])  # target: "bright"

# Model
model = Sequential([
    Embedding(input_dim=vocab_size, output_dim=10, input_length=4),
    LSTM(32),  # LSTM can capture long-term dependencies
    Dense(vocab_size, activation='softmax')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy')

# Train
model.fit(X, y, epochs=200, verbose=0)

# Predict next word
test = np.array([[0, 1, 2, 3]])  # "the sun is shining"
pred = model.predict(test)
print("Predicted next word:", idx2word[np.argmax(pred)])
```

# ENCODING AND DECODING

```python
from keras.models import Model
from keras.layers import Input, SimpleRNN, Dense
import numpy as np

input_text = "how are you"
target_text = "\tమీరు ఎలా ఉన్నారు\n"

input_chars = sorted(set(input_text))
target_chars = sorted(set(target_text))

input_idx = {ch: i for i, ch in enumerate(input_chars)}
target_idx = {ch: i for i, ch in enumerate(target_chars)}
reverse_target_idx = {i: ch for ch, i in target_idx.items()}

def one_hot(text, length, char_set):
    data = np.zeros((1, length, len(char_set)))
    for t, ch in enumerate(text):
        data[0, t, char_set[ch]] = 1
    return data

enc_input = one_hot(input_text, len(input_text), input_idx)

dec_input = one_hot(target_text[:-1], len(target_text)-1, target_idx)
dec_target = one_hot(target_text[1:], len(target_text)-1, target_idx)

enc_in = Input(shape=(None, len(input_idx)))
_, enc_state = SimpleRNN(64, return_state=True)(enc_in)

dec_in = Input(shape=(None, len(target_idx)))
dec_rnn = SimpleRNN(64, return_sequences=True, return_state=True)
dec_out, _ = dec_rnn(dec_in, initial_state=enc_state)
dec_dense = Dense(len(target_idx), activation='softmax')
dec_out = dec_dense(dec_out)

model = Model([enc_in, dec_in], dec_out)
model.compile(optimizer='adam', loss='categorical_crossentropy')
model.fit([enc_input, dec_input], dec_target, epochs=200, verbose=0)

enc_model = Model(enc_in, enc_state)

dec_state_input = Input(shape=(64,))
dec_single_input = Input(shape=(1, len(target_idx)))

dec_out2, dec_state2 = dec_rnn(dec_single_input, initial_state=dec_state_input)
dec_out2 = dec_dense(dec_out2)

dec_model = Model([dec_single_input, dec_state_input], [dec_out2, dec_state2])

state = enc_model.predict(enc_input)

target_seq = np.zeros((1,1,len(target_idx)))
target_seq[0,0,target_idx['\t']] = 1

decoded = ""

for _ in range(50):
    output_tokens, state = dec_model.predict([target_seq, state])
    idx = np.argmax(output_tokens[0,0])
    char = reverse_target_idx[idx]

    if char == '\n':
        break

    decoded += char

    target_seq = np.zeros((1,1,len(target_idx)))
    target_seq[0,0,idx] = 1

print("Input:", input_text)
print("Predicted:", decoded)
```

The implemented Encoder–Decoder model successfully performs sequence-to-sequence translation using SimpleRNN. The encoder converts the input sequence into a context vector, which is used by the decoder to generate the target sequence step-by-step. Teacher forcing is applied during training for faster convergence. Start and end tokens help control sequence generation. However, the model suffers from limitations such as small dataset size, lack of attention mechanism, and inability of SimpleRNN to capture long-term dependencies, leading to poor generalization.

Encoder Bottleneck Problem
All information stored in single vector

 Leads to:

Poor performance for longer sequences

 This is exactly why Attention was invented

```python
from keras.models import Model
from keras.layers import Input, SimpleRNN, Dense
import numpy as np

# Word pair for training
input_text = "how"
target_text = "ఎలా"

# Create character sets
input_chars = sorted(set(input_text))
target_chars = sorted(set(target_text))

input_idx = {ch: i for i, ch in enumerate(input_chars)}
target_idx = {ch: i for i, ch in enumerate(target_chars)}
reverse_target_idx = {i: ch for ch, i in target_idx.items()}

# One-hot encoding function
def one_hot(text, length, char_set):
    data = np.zeros((1, length, len(char_set)))
    for t, ch in enumerate(text):
        if ch in char_set:
            data[0, t, char_set[ch]] = 1
    return data

# Prepare training data
enc_input = one_hot(input_text, len(input_text), input_idx)
dec_input = one_hot(target_text, len(target_text), target_idx)
dec_target = np.copy(dec_input)

# Build the model
enc_in = Input(shape=(None, len(input_idx)))
_, enc_state = SimpleRNN(64, return_state=True)(enc_in)

dec_in = Input(shape=(None, len(target_idx)))
dec_out, _ = SimpleRNN(64, return_sequences=True, return_state=True)(dec_in, initial_state=enc_state)
dec_out = Dense(len(target_idx), activation='softmax')(dec_out)

model = Model([enc_in, dec_in], dec_out)
model.compile(optimizer='adam', loss='categorical_crossentropy')
model.fit([enc_input, dec_input], dec_target, epochs=300, verbose=0)

# Inference
test_word = "how"
test_input = one_hot(test_word, len(test_word), input_idx)
test_dec_input = np.zeros((1, len(target_text), len(target_idx)))  # just zeros for decoding
predicted = model.predict([test_input, test_dec_input])

# Decode prediction
translated = ''.join([reverse_target_idx[np.argmax(char)] for char in predicted[0]])
print(f"English: {test_word}")
print(f"Predicted Telugu: {translated}")
```

# GRU

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GRU, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Sample data
sentences = [
    "deep learning is powerful",
    "machine learning is fun",
    "deep learning is amazing"
]

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index

# Create sequences
input_sequences = []
for sentence in sentences:
    seq = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1, len(seq)):
        input_sequences.append(seq[:i+1])

# Pad sequences
max_len = max(len(x) for x in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')

# Split X and y
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

vocab_size = len(word_index) + 1

# Build GRU model
model = Sequential([
    Embedding(vocab_size, 10, input_length=max_len-1),
    GRU(64),
    Dense(vocab_size, activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')

# Train
model.fit(X, y, epochs=200, verbose=0)

# Predict next word
test = "deep learning is"
seq = tokenizer.texts_to_sequences([test])[0]
seq = pad_sequences([seq], maxlen=max_len-1, padding='pre')

pred = model.predict(seq)
pred_word = tokenizer.index_word[np.argmax(pred)]

print("Next word:", pred_word)
```

#Attention mechanism

```python
from keras.models import Model
from keras.layers import Input, LSTM, Dense, Attention
import numpy as np
from keras.layers import Concatenate

pairs = [
    ("how are you", "\tమీరు ఎలా ఉన్నారు\n"),
    ("good morning", "\tశుభోదయం\n"),
    ("what is your name", "\tమీ పేరు ఏమిటి\n"),
    ("thank you", "\tధన్యవాదాలు\n"),
]

input_texts = [p[0] for p in pairs]
target_texts = [p[1] for p in pairs]

input_chars = sorted(set(''.join(input_texts)))
target_chars = sorted(set(''.join(target_texts)))

input_idx = {ch: i for i, ch in enumerate(input_chars)}
target_idx = {ch: i for i, ch in enumerate(target_chars)}
reverse_target_idx = {i: ch for ch, i in target_idx.items()}
def encode_data(texts, char_idx, max_len):
    data = np.zeros((len(texts), max_len, len(char_idx)))
    for i, text in enumerate(texts):
        for t, ch in enumerate(text):
            data[i, t, char_idx[ch]] = 1
    return data

max_enc_len = max(len(t) for t in input_texts)
max_dec_len = max(len(t) for t in target_texts)

enc_input = encode_data(input_texts, input_idx, max_enc_len)
dec_input = encode_data([t[:-1] for t in target_texts], target_idx, max_dec_len-1)
dec_target = encode_data([t[1:] for t in target_texts], target_idx, max_dec_len-1)




#  Encoder
enc_in = Input(shape=(None, len(input_idx)))
enc_lstm = LSTM(64, return_sequences=True, return_state=True)
enc_out, state_h, state_c = enc_lstm(enc_in)

#  Decoder
dec_in = Input(shape=(None, len(target_idx)))
dec_lstm = LSTM(64, return_sequences=True, return_state=True)
dec_out, _, _ = dec_lstm(dec_in, initial_state=[state_h, state_c])

#  Attention
attn = Attention()
context = attn([dec_out, enc_out])



combined = Concatenate()([dec_out, context])
combined = Dense(64, activation='tanh')(combined)
# Final output
output = Dense(len(target_idx), activation='softmax')(combined)

model = Model([enc_in, dec_in], output)
model.compile(optimizer='adam', loss='categorical_crossentropy')

model.fit([enc_input, dec_input], dec_target, epochs=200, verbose=0)

#  Inference Models

enc_model = Model(enc_in, [enc_out, state_h, state_c])

# Decoder setup
dec_state_input_h = Input(shape=(64,))
dec_state_input_c = Input(shape=(64,))
enc_out_input = Input(shape=(None, 64))

dec_out2, state_h2, state_c2 = dec_lstm(
    dec_in, initial_state=[dec_state_input_h, dec_state_input_c]
)

context2 = attn([dec_out2, enc_out_input])
combined2 = Concatenate()([dec_out2, context2])
combined2 = Dense(64, activation='tanh')(combined2)
final_out = Dense(len(target_idx), activation='softmax')(combined2)

dec_model = Model(
    [dec_in, enc_out_input, dec_state_input_h, dec_state_input_c],
    [final_out, state_h2, state_c2]
)

#  Decoding
enc_out_val, h, c = enc_model.predict(enc_input)

target_seq = np.zeros((1,1,len(target_idx)))
target_seq[0,0,target_idx['\t']] = 1

decoded = ""

for _ in range(50):
    out, h, c = dec_model.predict([target_seq, enc_out_val, h, c])

    probs = out[0,0]
    idx = np.argmax(probs)
    char = reverse_target_idx[idx]

    if char == '\n':
        break

    decoded += char

    target_seq = np.zeros((1,1,len(target_idx)))
    target_seq[0,0,idx] = 1

print("Input:", input_text)
print("Predicted:", decoded)
```
