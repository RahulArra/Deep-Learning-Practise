```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()
x_train = x_train.reshape((60000, 28, 28, 1))
x_test = x_test.reshape((10000, 28, 28, 1))

x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),

    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

Conv2D(32) в†’ extracts basic features (edges, curves)

MaxPooling в†’ reduces spatial size ( it does not learn patterns)

Conv2D(64) в†’ learns more complex patterns

Flatten в†’ converts feature maps to vector

Dense(64) в†’ fully connected layer

Dense(10) в†’ output layer for digits 0вЂ“9

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)

print(test_acc)
```

we can observe the accuracy as 99 bcoz this is a simple dataset

```python
from tensorflow.keras.datasets import cifar10
# Load the CIFAR-10 dataset

(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python


model = models.Sequential([
      layers.Conv2D(64,(3,3),activation='relu',input_shape=(32,32,3)),
          layers.MaxPooling2D((2,2)),

              layers.Conv2D(128,(3,3),activation='relu'),
                  layers.MaxPooling2D((2,2)),

                      layers.Flatten(),

                          layers.Dense(128,activation='relu'),
                              layers.Dense(10,activation='softmax')
                              ])
```

 Train accuracy в‰€ 90%
Validation accuracy в‰€ 70
better feature learning
 higher training accuracy

But over fitting and more computation

```python
model = models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),

    layers.Dense(64,activation='relu'),
    layers.Dense(10,activation='softmax')
])
```

```python
model = models.Sequential([
    layers.Conv2D(64,(3,3),activation='relu',input_shape=(32,32,3)),
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.25),

    layers.Conv2D(128,(3,3),activation='relu'),
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.25),

    layers.Flatten(),

    layers.Dense(128,activation='relu'),
    layers.Dropout(0.5),

    layers.Dense(10,activation='softmax')
])
```

Dropout is applied during training
Dropout is NOT applied during validation/testing
So training get harder and testing use full network

Validation Loss в‰€ 1.14
After dropout:
Validation Loss в‰€ 0.80
Dropout reduces overfitting by randomly disabling neurons,
forcing the network to learn stronger and more general features.

```python
model = models.Sequential([
    layers.Conv2D(64,(3,3),padding='same',input_shape=(32,32,3)),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.25),

    layers.Conv2D(128,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.25),

    layers.Flatten(),

    layers.Dense(128),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Dropout(0.5),

    layers.Dense(10,activation='softmax')
])
```

Both training and validation accuracy improved.
BatchNorm stabilized the distribution of activations inside the network.
This allowed:
faster gradient flow
more stable updates
better feature learning

```python
model = models.Sequential([

    layers.Conv2D(64,(3,3),padding='same',input_shape=(32,32,3)),
    layers.BatchNormalization(),
    layers.Activation('relu'),

    layers.Conv2D(64,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),

    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.25),


    layers.Conv2D(128,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),

    layers.Conv2D(128,(3,3),padding='same'),
    layers.BatchNormalization(),
    layers.Activation('relu'),

    layers.MaxPooling2D((2,2)),
    layers.Dropout(0.25),


    layers.Flatten(),

    layers.Dense(256),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Dropout(0.5),

    layers.Dense(10,activation='softmax')
])
```

Increasing depth improved both training and validation performance.
Conv в†’ Conv в†’ Pool
Conv в†’ Conv в†’ Pool
This allowed the network to learn hierarchical features more effectively

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train,y_train,epochs=10,
          validation_data=(x_test,y_test))
```
